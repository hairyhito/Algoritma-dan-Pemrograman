Praktikum 04. Fungsi dengan *args dan **kwargs

```python
def penjumlahan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f" {key} : {value}")

print("Total:", penjumlahan(1,2,3,4,5))
biodata(nama="Andi", umur=20, prodi="TRK")
```

Praktikum 05. Scope Variabel

```python
x = 15

def ubah():
  x=5
  print("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

Praktikum 06. Fungsi Lambda

```python
kali = lambda x,y: x * y
print("hasil :", kali(5,4))
```

Praktikum 07. Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan ...")
    fungsi()
    print("Setelah fungsi dijalankan ...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Politeknik Negeri Semarang!")

sapa()
```

Praktikum 08. Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa

```python
def hitung_rata (data):
  return sum(data) / len(data)

nilai= [80,90,95]
rata = hitung_rata (nilai)
print("Rata-rata nilai:", rata)
```

Penugasan 08. Fungsi Decorator

```python
def decorator_logger(halo):
  def wrapper():
    print("Memulai fungsi ...")
    halo()
    print("Fungsi selesai dijalankan ...")
  return wrapper

@decorator_logger
def halo():
  print("Halo, Selamat siang!")

halo()
```

Penugasan 09. Studi Kasus Nyata - Program Nilai

```python
def input_nilai():
    tugas = float(input("Masukkan nilai tugas: "))
    ujian = float(input("Masukkan nilai ujian: "))
    return [tugas, ujian]

def hitung_rata(data, fungsi_lambda):
    return fungsi_lambda(data[0], data[1])

def status_kelulusan(rata):
    if rata >= 60:
        return "LULUS"
    else:
        return "TIDAK LULUS"

hitung_bobot = lambda t, u: (t * 0.40) + (u * 0.60)
hitung_rata_sederhana = lambda t, u: (t + u) / 2

data_mahasiswa = input_nilai()

ratarata = hitung_rata(data_mahasiswa, hitung_rata_sederhana)
nilai_akhir = hitung_rata(data_mahasiswa, hitung_bobot)
status = status_kelulusan(nilai_akhir)

print(f"Rata rata: {ratarata}")
print(f"Nilai Akhir: {nilai_akhir}")
print(f"Status: {status}")
```
