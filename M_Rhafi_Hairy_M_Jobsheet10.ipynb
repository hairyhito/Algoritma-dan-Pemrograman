# Praktikum

Praktikum 01. Bubble Sort

```python
def bubble_sort(data):
      n = len(data)
      for i in range(n):
        for j in range(0, n - i - 1):
          if data[j] > data[j + 1]:
           data[j], data[j + 1] = data[j + 1], data[j]
      return data
angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

Praktikum 02. Selection Sort


```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
            data[i], data[min_index] = data[min_index], data[i]
        return data
angka = [64, 34, 25, 12, 22, 11,]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

Praktikum 03. Insertion Sort

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

Praktikum 04. Studi Kasus Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] < data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

Praktikum 05. Perbandingan Waktu Eksekusi

```python
import random, time

data = [random.randint(1, 1000) for _ in range(500)]

for func in [bubble_sort, selection_sort, insertion_sort]:
  salinan = data.copy()
  start = time.time()
  func(salinan)
  end = time.time()
  print(f"{func.__name__} membutuhkan waktu {end - start:.5f} detik")
```

Praktikum 06. Sorting Data Barang

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen",
10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

# Penugasan


Penugasan 01. Eksperimen Bubble Sort

```python
def bubble_sort_descending(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data awal:", angka)
print("Data descending:", bubble_sort_descending(angka))
```

Penugasan 02. Eksperimen Selection Sort

```python
def selection_sort_max_to_end(data):
    n = len(data)
    for i in range(n - 1, 0, -1):
        max_index = i
        for j in range(i):
            if data[j] > data[max_index]:
                max_index = j
        data[i], data[max_index] = data[max_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data awal:", angka)
print("Data terurut:", selection_sort_max_to_end(angka))
```

Penugasan 03. Eksperimen Insertion Sort

```python
def insertion_sort_strings(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

nama_mahasiswa = ["Zara", "Ali", "Dina", "Budi", "Citra"]
print("Nama awal:", nama_mahasiswa)
print("Nama terurut:", insertion_sort_strings(nama_mahasiswa))
```

Penugasan 05. Analisis Kompleksitas

```python
import random

def count_bubble_sort(data):
    count = 0
    n = len(data)
    arr = data.copy()
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return count

def count_selection_sort(data):
    count = 0
    n = len(data)
    arr = data.copy()
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            count += 1
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return count

def count_insertion_sort(data):
    count = 0
    n = len(data)
    arr = data.copy()
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0:
            count += 1
            if arr[j] > key:
                arr[j + 1] = arr[j]
                j -= 1
            else:
                break
        arr[j + 1] = key
    return count

sizes = [10, 50, 100]
print(f"{'Size':<10} | {'Bubble':<10} | {'Selection':<10} | {'Insertion':<10}")
print("-" * 50)

for size in sizes:
    data_acak = [random.randint(1, 1000) for _ in range(size)]

    b_count = count_bubble_sort(data_acak)
    s_count = count_selection_sort(data_acak)
    i_count = count_insertion_sort(data_acak)

    print(f"{size:<10} | {b_count:<10} | {s_count:<10} | {i_count:<10}")
```

Penugasan 06. Proyek Mini

```python
def bubble_sort_barang(data, key_index, is_descending):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            condition = False
            if is_descending:
                if data[j][key_index] < data[j + 1][key_index]:
                    condition = True
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    condition = True

            if condition:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

daftar_barang = []
jumlah = int(input("Masukkan jumlah barang: "))

for i in range(jumlah):
    print(f"Barang ke-{i+1}")
    nama = input("Nama Barang: ")
    harga = int(input("Harga Barang: "))
    daftar_barang.append((nama, harga))

print("\n--- Menu Sorting ---")
print("1. Urutkan berdasarkan Nama (A-Z)")
print("2. Urutkan berdasarkan Nama (Z-A)")
print("3. Urutkan berdasarkan Harga (Termurah)")
print("4. Urutkan berdasarkan Harga (Termahal)")
pilihan = input("Pilih menu (1-4): ")

if pilihan == '1':
    hasil = bubble_sort_barang(daftar_barang, 0, False)
elif pilihan == '2':
    hasil = bubble_sort_barang(daftar_barang, 0, True)
elif pilihan == '3':
    hasil = bubble_sort_barang(daftar_barang, 1, False)
elif pilihan == '4':
    hasil = bubble_sort_barang(daftar_barang, 1, True)
else:
    hasil = daftar_barang
    print("Pilihan tidak valid")

print("\nHasil Pengurutan:")
for item in hasil:
    print(f"{item[0]} - Rp {item[1]}")
```
