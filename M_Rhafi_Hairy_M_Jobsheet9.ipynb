Penugasan 06. Studi Kasus Mini

```python
%%writefile laporan.py
def tampilkan_skor(benar, salah, total_soal):
    persentase = (benar / total_soal) * 100
    print("\n" + "="*30)
    print("      RINGKASAN SKOR KUIS")
    print("="*30)
    print(f"Total Soal yang Dikerjakan: {total_soal}")
    print(f"Jawaban Benar: {benar}")
    print(f"Jawaban Salah: {salah}")
    print("-" * 30)
    print(f"Persentase Benar: {persentase:.2f}%")
    print("="*30)
```

```python
import random
import laporan

skor_benar = 0
skor_salah = 0
jumlah_soal = 3

print("=== SELAMAT DATANG DI KUIS PENJUMLAHAN ===")

for i in range(1, jumlah_soal + 1):
    angka1 = random.randint(1, 20)
    angka2 = random.randint(1, 20)
    jawaban_sebenarnya = angka1 + angka2

    soal = f"Soal ke-{i}: Berapakah hasil dari {angka1} + {angka2}? "

    while True:
        try:
            jawaban_user = int(input(soal))
            break
        except ValueError:
            print("Input tidak valid. Harap masukkan angka.")

    if jawaban_user == jawaban_sebenarnya:
        print("Jawaban Benar!")
        skor_benar += 1
    else:
        print(f"Jawaban Salah. Jawaban yang benar adalah {jawaban_sebenarnya}.")
        skor_salah += 1


laporan.tampilkan_skor(skor_benar, skor_salah, jumlah_soal)
```
